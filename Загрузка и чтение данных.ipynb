**1.2. Изучение структуры и типов данных**

```python
# Схема данных
import pandas as pd

schema = pd.read_csv("schema.csv")
print("Размер schema:", schema.shape)
display(schema)
```

```python
# Схема vmtable
schema_vmtable = schema[schema["filepattern"].str.contains("vmtable", na=False)]
display(schema_vmtable)
```

```python
# Схема vm_cpu_readings
schema_cpu = schema[schema["filepattern"].str.contains("vm_cpu_readings", na=False)]
display(schema_cpu)
```

```python
# Структура vmtable
vmtable = pd.read_csv("vmtable.csv.gz", compression="gzip")

print("Размер vmtable:", vmtable.shape)
print("\nКолонки vmtable:")
print(vmtable.columns.tolist())

display(vmtable.head())
```

```python
# Типы данных и пропуски в vmtable
vmtable.info()

print("\nДоля пропусков в vmtable:")
display(vmtable.isna().mean().sort_values(ascending=False))
```

```python
# Структура CPU-файлов
cpu1 = pd.read_csv(
    "vm_cpu_readings-file-1-of-195.csv.gz",
    compression="gzip"
)

print("Размер cpu1:", cpu1.shape)
print("\nКолонки cpu1:")
print(cpu1.columns.tolist())

display(cpu1.head())
```

```python
# Типы данных и пропуски в cpu1
cpu1.info()

print("\nДоля пропусков в cpu1:")
display(cpu1.isna().mean().sort_values(ascending=False))
```

```python
print("Колонки vmtable:", vmtable.columns.tolist())
print("Колонки cpu1:", cpu1.columns.tolist())
```

```python
#JOIN
merged = cpu1.merge(vmtable, on="vmid", how="left")

print(merged.shape)
merged.head()
```
