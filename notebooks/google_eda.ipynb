```python
usage_info = pd.DataFrame({
    "table": "instance_usage",
    "column": usage.columns,
    "dtype": usage.dtypes.astype(str).values,
    "non_null_count": usage.notnull().sum().values
})
```

```python
events_info = pd.DataFrame({
    "table": "instance_events",
    "column": events.columns,
    "dtype": events.dtypes.astype(str).values,
    "non_null_count": events.notnull().sum().values
})
```

```python
data_dictionary = pd.concat([usage_info, events_info], ignore_index=True)
data_dictionary
```
